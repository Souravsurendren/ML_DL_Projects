```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
```

```python
xx = dataframe.drop('Aggregate rating', axis=1)
yy = dataframe['Aggregate rating']
```

```python
xx_train,xx_test,yy_train,yy_test=train_test_split(xx,yy,test_size=.1,random_state=353)
xx_train.head()
yy_train.head()
```

Linear Regression

```python
regre=LinearRegression()
regre.fit(xx_train,yy_train)
yy_pred=regre.predict(xx_test)
from sklearn.metrics import r2_score
r2_score(yy_test,yy_pred)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
regre = LinearRegression()
regre.fit(xx_train, yy_train)
yy_pred = regre.predict(xx_test)
mse = mean_squared_error(yy_test, yy_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
r2 = r2_score(yy_test, yy_pred)
print(f"R-squared (R2) Error: {r2:.2f}")
```

Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
xx_train,xx_test,yy_train,yy_test=train_test_split(xx,yy,test_size=.1,random_state=105)
DTree=DecisionTreeRegressor(min_samples_leaf=.0001)
DTree.fit(xx_train,yy_train)
yy_predict=DTree.predict(xx_test)
from sklearn.metrics import r2_score
r2_score(yy_test,yy_predict)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

DTree = DecisionTreeRegressor(min_samples_leaf=0.0001)

DTree.fit(xx_train, yy_train)

yy_predict = DTree.predict(xx_test)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(yy_test, yy_predict)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Calculate R-squared (R2) Error
r2 = r2_score(yy_test, yy_predict)
print(f"R-squared (R2) Error: {r2:.2f}")
```
